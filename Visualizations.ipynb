```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [2,4,6]
```

```python
x2 = [4,5,6]
y2 = [5,7,9]
```

```python
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Sample Graph")
```

```python
# Normal dotted plot.
plt.plot(x, y, label="First Curve")
plt.plot([4,5,6], [7,8,9], label="Second Curve")
plt.legend()
plt.show()
```

```python
# Bar charts.
plt.title("Bar chart!")
plt.bar(x, y, label="Bars 1", color='c')
plt.bar(x2, y2, label="Bars 2", color='r')
plt.legend()
plt.show()
```

```python
# Histograms
ages = [10,45,30,60,35,67,89,99,35,10, 78]
bins = [10, 20, 30, 40, 50, 60, 80, 80, 90, 100]

plt.xlabel("Ages")
plt.ylabel("Count")
plt.hist(ages, bins, histtype='bar', rwidth=0.8, color='c')
plt.show()
```

```python
# Scatter plot
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter Plot")
plt.scatter(x, y, label="Scatter Plot", color='k', marker='*')
plt.legend()
plt.show()
```

```python
# Stack plot
days = [1,2,3,4,5,6,7]
sleeping = [ 6, 8, 9, 5, 4,10,10]
eating =   [ 1, 2, 2, 1, 1, 2, 2]
working =  [14,12,10,14,14, 5, 4]
playing =  [ 3, 2, 3, 4, 5, 7, 8]


plt.stackplot(days, sleeping, eating, working, playing, colors = ['m', 'c', 'r', 'g'])
plt.xlabel("Days")
plt.ylabel("Activity")
plt.title("Time distribution")
# Empty plots to workaround lack of labels on stack plot.
plt.plot([], [], label='Sleeping', color = 'm', linewidth=3)
plt.plot([], [], label='Eating', color = 'c', linewidth=3)
plt.plot([], [], label='Working', color = 'r', linewidth=3)
plt.plot([], [], label='Playing', color = 'g', linewidth=3)
plt.legend()
plt.show()
```

```python
# Pie chart

slices = [7, 2, 2, 13]
activities = ['Sleeping', 'Eating', 'Working', 'Playing']
cols = ['c', 'm', 'r', 'b']
plt.pie(slices, 
        labels=activities, 
        colors=cols, 
        startangle=90, 
        shadow=False, 
        explode=(0, 0.1, 0, 0), 
        autopct="%1.1%%f")
plt.show()
```

```python
# Reading from internet
from pandas_datareader import data, wb

def graph_data(stock):
    df = data.DataReader(stock,'yahoo')
    plt.plot_date(df.index, df.Close, '-')
    
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title('Yahoo Stock')
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()

graph_data('TSLA')
```

```python
# Using ChartDirectory library for generating candlestick based visualizations.

from pychartdir import *

highData = [2043, 2039, 2076, 2064, 2048, 2058, 2070, 2033, 2027, 2029, 2071, 2085, 2034, 2031,
    2056, 2128, 2180, 2183, 2192, 2213, 2230, 2281, 2272]

lowData = [1931, 1921, 1985, 2028, 1986, 1994, 1999, 1958, 1943, 1944, 1962, 2011, 1975, 1962, 1928,
    2059, 2112, 2103, 2151, 2127, 2123, 2152, 2212]

openData = [2000, 1957, 1993, 2037, 2018, 2021, 2045, 2009, 1959, 1985, 2008, 2048, 2006, 2010,
    1971, 2080, 2116, 2137, 2170, 2172, 2171, 2191, 2240]

closeData = [1950, 1991, 2026, 2029, 2004, 2053, 2011, 1962, 1987, 2019, 2040, 2016, 1996, 1985,
    2006, 2113, 2142, 2167, 2158, 2201, 2188, 2231, 2242]

labels = ["Mon 1", "Tue 2", "Wed 3", "Thu 4", "Fri 5", "Mon 8", "Tue 9", "Wed 10", "Thu 11",
    "Fri 12", "Mon 15", "Tue 16", "Wed 17", "Thu 18", "Fri 19", "Mon 22", "Tue 23", "Wed 24",
    "Thu 25", "Fri 26", "Mon 29", "Tue 30", "Wed 31"]

c = XYChart(600, 350)

# Set the plotarea at (50, 25) and of size 500 x 250 pixels. Enable both the horizontal and vertical
# grids by setting their colors to grey (0xc0c0c0)
c.setPlotArea(50, 25, 500, 250).setGridColor(0xc0c0c0, 0xc0c0c0)

# Add a title to the chart
c.addTitle("Universal Stock Index on Jan 2001")

# Add a custom text at (50, 25) (the upper left corner of the plotarea). Use 12pt Arial Bold/blue
# (4040c0) as the font.
c.addText(50, 25, "(c) Global XYZ ABC Company", "arialbd.ttf", 12, 0x4040c0)

# Add a title to the x axis
c.xAxis().setTitle("Jan 2001")

# Set the labels on the x axis. Rotate the labels by 45 degrees.
c.xAxis().setLabels(labels).setFontAngle(45)

# Add a title to the y axis
c.yAxis().setTitle("Universal Stock Index")

# Draw the y axis on the right hand side of the plot area
c.setYAxisOnRight(1)

# Add a CandleStick layer to the chart using green (00ff00) for up candles and red (ff0000) for down
# candles
layer = c.addCandleStickLayer(highData, lowData, openData, closeData, 0x00ff00, 0xff0000)

# Set the line width to 2 pixels
layer.setLineWidth(2)

# Output the chart
c.makeChart("candlestick.png")
```
